```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "C:/Users/powde/Desktop/pandas-challege/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
# Display the total number of players
player_count = purchase_data["SN"].nunique()
pc_df = pd.DataFrame({"Total Players": [player_count]})
pc_df
```

```python
#Purchasing Analysis (Total)
#unique Item count
unique_items = purchase_data["Item ID"].nunique()
unique_items

#Average price
avg_price = round(purchase_data["Price"].mean(), 2)

#Number of Purchases
purchase_cnt = purchase_data["Purchase ID"].count()
purchase_cnt

#Total Revenue
total_revenue = purchase_data["Price"].sum()

#Summary table
summary_table = pd.DataFrame({"Number of Unique Items": unique_items,
                              "Average Price" : [avg_price],
                              "Number of Purchases" : [purchase_cnt],
                              "Total Revenue": [total_revenue]})

#Summary table formatting
summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:.2f}".format)
summary_table["Average Price"] = summary_table["Average Price"].map("${:.2f}".format)
summary_table
```

```python
#Gender Demos
purchase_data_grouped = purchase_data.groupby("SN").first()

#counts
male_cnt = purchase_data_grouped["Gender"].value_counts()["Male"]
female_cnt = purchase_data_grouped["Gender"].value_counts()["Female"]
other_ndc_cnt = purchase_data_grouped["Gender"].value_counts()["Other / Non-Disclosed"]

#Total
total_cnt = male_cnt + female_cnt + other_ndc_cnt 

gender_list = purchase_data_grouped["Gender"].unique()

gender_cnt_all = (male_cnt, female_cnt, other_ndc_cnt)
pct_male = male_cnt / total_cnt * 100
pct_female = female_cnt / total_cnt * 100
pct_other_ndc = other_ndc_cnt / total_cnt * 100

percentages = (pct_male, pct_female, pct_other_ndc)

gender_table = pd.DataFrame({"Gender": gender_list,
                            "Total Count": gender_cnt_all,
                            "Percentage of Players": percentages})

gender_table["Percentage of Players"] = gender_table["Percentage of Players"].astype(float).map("{:.2f}%".format)

gender_table

```

```python
 #Purchasing Analysis (Gender)
gender_df = purchase_data.groupby(["Gender"])

purchase_count = gender_df["Price"].count()
price_mean = gender_df["Price"].mean()
total_purchase = gender_df["Price"].sum()
ave_tot_per = gender_df["Price"].sum() / gender_df["SN"].nunique()

gender_purchasing = pd.DataFrame({"Purchase Count": purchase_count,
                                  "Average Purchase Price": price_mean,
                                  "Total Purchase Value": total_purchase,
                                  "Avg Total Purchase per Person" : ave_tot_per
                                   })

gender_purchasing["Average Purchase Price"] = gender_purchasing["Average Purchase Price"].astype(float).map("${:,.2f}".format)
gender_purchasing["Total Purchase Value"] = gender_purchasing["Total Purchase Value"].astype(float).map("${:,.2f}".format)
gender_purchasing["Avg Total Purchase per Person"] = gender_purchasing["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

gender_purchasing
```

```python
#Age Demographics
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
#purchase_data

age_ranges_df = purchase_data.groupby(["Age Range"])
#age_ranges_df 

player_count_by_age = age_ranges_df["SN"].nunique()
pct_players_by_age = (age_ranges_df["SN"].nunique() / total_cnt) * 100
player_count_by_age
pct_players_by_age

age_demographics_df = pd.DataFrame({"Total Count": player_count_by_age,
                                    "Percentage of Players": pct_players_by_age
                                    })

age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].astype(float).map("{:.2f}%".format)
age_demographics_df["Total Count"] = age_demographics_df["Total Count"].astype(float).map("{:,.0f}".format)


age_demographics_df
```

```python
 #Purchasing Analysis (Age)
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


age_purchase_grouped = purchase_data.groupby(["Age Range"])
#age_purchase_grouped

purchase_cnt = age_purchase_grouped["SN"].count()
#purchase_cnt

avg_purchase_price = age_purchase_grouped["Price"].mean()
purchase_value_total = age_purchase_grouped["Price"].sum()

ave_purchase_price = age_purchase_grouped["Price"].sum() / age_purchase_grouped["SN"].nunique()



#create df
age_purchase_df = pd.DataFrame({"Purchase Count": purchase_cnt,
                                "Average Purchase Price": avg_purchase_price,
                                "Total Purchase Value" : purchase_value_total,
                                "Avg Total Purchase per Person" : ave_purchase_price})
#Format string
age_purchase_df["Average Purchase Price"] = age_purchase_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_purchase_df["Total Purchase Value"] = age_purchase_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_purchase_df["Avg Total Purchase per Person"] = age_purchase_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
age_purchase_df
```

```python
#Top Spenders
#group by SN
sn_purchase_grouped = purchase_data.groupby(["SN"])

#Get calc column values
sn_purch_cnt = sn_purchase_grouped["SN"].count()
sn_purch_cnt = sn_purch_cnt.sort_values(ascending=False)
#sn_purch_cnt.head(10)

avg_purch_price = sn_purchase_grouped["Price"].mean()
purch_value_total = sn_purchase_grouped["Price"].sum()

#Dataframe
purch_analysis_df = pd.DataFrame({"Purchase Count" : sn_purch_cnt,
                                  "Average Purchase Price" : avg_purch_price,
                                  "Total Purchase Value" : purch_value_total})
#Sorting
purch_analysis_df = purch_analysis_df.sort_values(["Total Purchase Value"], ascending=[False])

#Formatting
purch_analysis_df["Average Purchase Price"] = purch_analysis_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purch_analysis_df["Total Purchase Value"] = purch_analysis_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purch_analysis_df.head(10)
```

```python
#Most Popular Items
#group by item
purchase_data.columns
popular_items_grouped = purchase_data.groupby(["Item ID","Item Name"])

purch_cnt = popular_items_grouped["Item Name"].count()
#purch_cnt.head(5)
unit_price = popular_items_grouped["Price"].mean()
#unit_price
total_price = popular_items_grouped["Price"].sum()

most_popular_df = pd.DataFrame({"Purchase Count" : purch_cnt,
                                "Item Price" : unit_price,
                                "Total Purchase Value" : total_price})

most_popular_df["Item Price"] = most_popular_df["Item Price"].astype(float).map("${:,.2f}".format)
most_popular_df["Total Purchase Value"] = most_popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
most_popular_df = most_popular_df.sort_values("Purchase Count", ascending=False)
most_popular_df.head(10)
```

```python
 #Most Profitable Items
most_profitable_df = pd.DataFrame({"Purchase Count" : purch_cnt,
                                "Item Price" : unit_price,
                                "Total Purchase Value" : total_price})

#Sorting
most_profitable_df = most_profitable_df.sort_values("Total Purchase Value", ascending=False)

#Formatting
most_profitable_df["Item Price"] = most_profitable_df["Item Price"].astype(float).map("${:,.2f}".format)
most_profitable_df["Total Purchase Value"] = most_profitable_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
most_profitable_df.head(10)
```
